```python
drop_cols = ['GINI index (World Bank estimate)', 
                'GINI index (World Bank estimate), average 2000-15',
                'gini of household income reported in Gallup, by wp5-year',
                'Democratic Quality',
                'Delivery Quality']

new_df = df.drop(columns = drop_cols)
new_df
```

```python
impute_cols = ['Log GDP per capita', 'Social support', 'Healthy life expectancy at birth',
                  'Freedom to make life choices', 'Generosity', 'Perceptions of corruption',
                  'Positive affect', 'Negative affect', 'Confidence in national government']


imputer = SimpleImputer(strategy = 'median')
new_df[impute_cols] = imputer.fit_transform(new_df[impute_cols])
new_df
```

```python
new_df.isna().sum()
```

```python
skewness = new_df[features].skew()
print(skewness)
print("Life Ladder skew:", new_df[label].skew()) # -> no need to winsorize and from the boxplots it can be seen that there are no extreme outliers
```

```python
new_features = []
def winsorize_column(df, col):
    lower_bound = new_df[col].quantile(0.01)
    upper_bound = new_df[col].quantile(0.99)
    new_col = f"label_{col.replace(' ', '_')}"
    new_features.append(new_col)
    
    new_df[new_col] = np.where(new_df[col] < lower_bound, lower_bound,
                           np.where(new_df[col] > upper_bound, upper_bound, new_df[col]))
    return new_df

for feature in features:
    new_df = winsorize_column(new_df, feature)
```

```python
visualize_outliers(new_features)
```

```python
scaler = StandardScaler()
final_features = scaler.fit_transform(new_df[new_features])
```

```python
X = final_features
y = new_df[label]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1234)
```

```python
# Train and evaluate Linear Regression
lr = LinearRegression()

lr_cv_scores_r2 = cross_val_score(lr, final_features, new_df[label].values, cv=5, scoring='r2')
print("Linear Regression CV R^2 Scores:", lr_cv_scores_r2)
print("Mean CV R^2:", lr_cv_scores_r2.mean())

lr_cv_scores_mse = cross_val_score(lr, final_features, new_df[label].values, cv=5, scoring='neg_mean_squared_error')
lr_mse_scores = -lr_cv_scores_mse  # negate to get actual MSE
print("Linear Regression CV MSE Scores:", lr_mse_scores)
print("Mean CV MSE:", lr_mse_scores.mean())
```

```python
#hyperparameter range for ridge and lasso 
# alpha -> penalty term
param_grid = {
    "alpha": [0.01, 0.01, 1, 10, 100, 1000]
}
```

```python
# Perform Ridge Regression
ridge = Ridge()
ridge_cv = GridSearchCV(ridge, param_grid, cv=5, scoring='r2')
ridge_cv.fit(X_train, y_train)

ridge_best_alpha = ridge_cv.best_params_['alpha']
print(f"Best Ridge alpha: {ridge_best_alpha} with CV R^2: {ridge_cv.best_score_:.4f}")

# Evaluate on test set
ridge_best = ridge_cv.best_estimator_
y_pred_ridge = ridge_best.predict(X_test)
print("Ridge Regression R^2:", r2_score(y_test, y_pred_ridge))
print("Ridge Regression MSE:", mean_squared_error(y_test, y_pred_ridge))
```

```python
#Perform Lasso Regression
lasso = Lasso()
lasso_cv = GridSearchCV(lasso, param_grid, cv=5, scoring='r2')
lasso_cv.fit(X_train, y_train)

lasso_best_alpha = lasso_cv.best_params_['alpha']
print(f"Best Lasso alpha: {lasso_best_alpha} with CV R^2: {lasso_cv.best_score_:.4f}")

# Evaluate on test set
lasso_best = lasso_cv.best_estimator_
y_pred_lasso = lasso_best.predict(X_test)
print("Lasso Regression R^2:", r2_score(y_test, y_pred_lasso))
print("Lasso Regression MSE:", mean_squared_error(y_test, y_pred_lasso))
```

```python
# final model
best_alpha = ridge_best_alpha
final_model = Ridge(alpha=best_alpha)
final_model.fit(X_train, y_train)

y_pred_final = final_model.predict(X_test)
r2_best = r2_score(y_test, y_pred_final)
mse_best = mean_squared_error(y_test, y_pred_final)

print("Best Model: Ridge Regression R:", r2_best)
print("Best Model: Ridge Regression MSE:", mse_best)
```

```python
# Step 5: Model optimization
alphas = np.logspace(0, 1, 1000)

new_ridge = Ridge()
new_ridge_cv = GridSearchCV(estimator=new_ridge, param_grid={'alpha': alphas}, cv=5, scoring='r2', n_jobs=-1)
new_ridge_cv.fit(X_train, y_train)

new_ridge_best_alpha = new_ridge_cv.best_params_['alpha']
print(f"Best Ridge alpha: {new_ridge_best_alpha} with CV R^2: {new_ridge_cv.best_score_:.4f}")

# Evaluate on test set
new_ridge_best = new_ridge_cv.best_estimator_
y_pred_new_ridge = new_ridge_best.predict(X_test)
print("Ridge Regression R^2:", r2_score(y_test, y_pred_new_ridge))
print("Ridge Regression MSE:", mean_squared_error(y_test, y_pred_new_ridge))
```

We observe no change in metrics from the best model achieved previously. 

```python
# Attempting non-linear models to capture non-linear relationships
# Random Forest

rf_param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [None, 10, 20, 50],
    'min_samples_leaf': [1, 2, 5]
}

rf = RandomForestRegressor(random_state=42)
rf_cv = GridSearchCV(rf, param_grid=rf_param_grid, cv=5, scoring='r2')

print("Tuning Random Forest...")
rf_cv.fit(X_train, y_train)

rf_best_params = rf_cv.best_params_
print(f"Best Random Forest params: {rf_best_params}")
print(f"Best Random Forest CV R^2: {rf_cv.best_score_:.4f}")

rf_best = rf_cv.best_estimator_
y_pred_rf = rf_best.predict(X_test)
r2_score_rf = r2_score(y_test, y_pred_rf)
mse_score_rf = mean_squared_error(y_test, y_pred_rf)
print(f"Random Forest Test R^2:", r2_score_rf)
print(f"Random Forest Test MSE:", mse_score_rf)
```

As seen above, by applying a non-linear regressor, we've achieved a higher accuracy of 0.876 increasing by about 0.1(12% increase). Random Forest also allowed us to decrease our MSE from 0.282 to 0.162.
